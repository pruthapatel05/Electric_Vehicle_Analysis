```python
# TASK 4: Sentiment Analysis with Visualization

from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
# Step 1: Sample Amazon reviews
reviews = [
    "This product is amazing! Totally worth the money.",
    "The item was okay, not great but not terrible either.",
    "Terrible quality, broke after one day. Very disappointed.",
    "Fast delivery and great packaging!",
    "I expected better performance for the price.",
    "Absolutely love it! Will buy again.",
    "Not as described. Waste of money.",
    "Decent quality for the price range.",
    "Customer service was super helpful and kind.",
    "The color faded after a week of use."
]
```

```python
# Step 2: Initialize counters
positive, neutral, negative = 0, 0, 0
total_score = 0
```

```python
# Step 3: Analyze each review
for review in reviews:
    blob = TextBlob(review)
    score = blob.sentiment.polarity  # -1 (negative) to +1 (positive)
    total_score += score

    if score > 0.1:
        sentiment = "Positive"
        positive += 1
    elif score < -0.1:
        sentiment = "Negative"
        negative += 1
    else:
        sentiment = "Neutral"
        neutral += 1
```

```python
print("Amazon Review Sentiment Analysis\n" + "-"*45)

# Step 4: Summary
print("\n" + "="*45)
print("Sentiment Summary")
print(f"Positive Reviews: {positive}")
print(f"Neutral Reviews : {neutral}")
print(f"Negative Reviews: {negative}")
print(f"Average Sentiment Score: {total_score/len(reviews):.2f}")
```

```python
# Step 5: Insights
print("\n Insights:")
if positive > negative:
    print("Most customers are happy with the product — good sign for marketing!")
elif negative > positive:
    print("Many customers are dissatisfied — improvements might be needed.")
else:
    print("Mixed feedback — product performance is average.")
```

```python
# Step 6: Visualization (Bar Chart)
labels = ['Positive', 'Neutral', 'Negative']
counts = [positive, neutral, negative]
colors = ['green', 'gold', 'red']
plt.figure(figsize=(7,5))
plt.bar(labels, counts, color=colors)
plt.title("Amazon Reviews Sentiment Analysis", fontsize=14, fontweight='bold')
plt.xlabel("Sentiment Type")
plt.ylabel("Number of Reviews")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
